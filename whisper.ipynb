Voice Recording

```python
import sounddevice as sd
import numpy as np
import threading
import queue
import time
from scipy.io.wavfile import write
```

```python
fs=16000 
channels=1
filename="mic_dynamic.wav"
```

```python
q=queue.Queue()
recording=[]
recording_active = True
```

```python
def audio_callback(indata, frames, time_info, status):
    if status:
        print(status)
    q.put(indata.copy())

def record_audio():
    with sd.InputStream(samplerate=fs, channels=channels, callback=audio_callback):
        print("The recording has started. Click on OK to stop it.")
        while recording_active:
            block = q.get()
            recording.append(block)
```

```python
thread = threading.Thread(target=record_audio)
thread.start()
```

```python
input(" Click on OK to stop recording ")
recording_active = False
thread.join()
```

```python
audio_data = np.concatenate(recording, axis=0)
write(filename, fs, audio_data)
```

```python
print(f"The record has taken successfully and saved as '{filename}'.")
```

```python
def extract_text_from_srt(srt_path):  #srt file
    with open(srt_path, "r", encoding="utf-8") as f:
        lines = f.readlines()

    transcript_lines = []
    for line in lines:
        line = line.strip()
        if line and not line.isdigit() and "-->" not in line:
            transcript_lines.append(line)

    return " ".join(transcript_lines)
```
